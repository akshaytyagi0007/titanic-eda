# Titanic Survival EDA
Exploratory Data Analysis on the Titanic dataset using Seaborn and Matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv(filepath_or_buffer="train.csv")
```

## 1. Survival Rate by Gender

```python
fig = plt.figure(figsize=(5, 3))
sns.barplot(data=dataset, x="Sex", y="Survived")
plt.title("Survival Rate by Gender")
plt.show()
```

## 2. Survival Rate by Passenger Class

```python
sns.barplot(data=dataset, x="Pclass", y="Survived")
plt.title("Survival Rate By Passenger Class")
plt.show()
```

## 3. Survival Rate by Age

```python
sns.histplot(data=dataset, x="Age", hue="Survived", bins=10, multiple="stack")
plt.title("Survival Rate By Age")
plt.show()
```

## 4. Fare Distribution by Passenger Class and Survival

```python
sns.boxplot(data=dataset, x="Pclass", y="Fare", hue="Survived")
plt.title("Fare Distribution by Class and Survival")
plt.show()
```

## 5. Survival Rate by Family Size

```python
dataset["familysize"] = dataset["SibSp"] + dataset["Parch"] + 1
sns.barplot(data=dataset, x="familysize", y="Survived")
plt.title("Survival Rate by Family Size")
plt.show()
```
